The Shaolin DataFrame scatter plot allows to create a fully customizable scatter plot from an arbitrary pandas DataFrame. We can not only select which pair of columns will be plotted but also we have full control over all configurable plot parameters using widgets.

In order to use it you just need to import the DataFrameScatter and initialize it with the DataFrame you want to plot.

```python
from IPython.display import Image #this is for displaying the widgets in the web version of the notebook
from shaolin.benders.pandas import DataFrameScatter
from bokeh.sampledata.iris import flowers
ds = DataFrameScatter(flowers)
```

## Displaying the plot

There are several ways of displaying the scatter plot. Displaying a plot with the full interface can be done through the widget attribute as shown below. 

It is also possible to disable individual component using the toggle buttons located below the control panel.

```python
#Image(filename='scatter_data/widget.png') 
ds.widget
```

## Displaying the plot in a sepparate cell

It is possible to display each individual element of the control panel separatedly. For example now we will be displaying just the Mapper and the plot in two sepparate cells. 

The plot is located as the "plot" attribute of the DataFrameScatter. Using the show function from bokeh it can be displayed as an standalone plot.



```python
from bokeh.plotting import show
```

```python
_ = show(ds.plot)
```

## Parameter Mapper

The parameter mapper (MarkerMappedParams class) is an object in charge of mapping the columns of the input DataFrame to any of the parameters that an scatter plot marker has. 

The selection of a column for the x and y parameters will be always active as they are needed for displaying the plot. 

Additional attributes can be mapped to a column bu clicking on the checkbox that every attribute has. 

The checkbox located near the x attribute allow to change all the other column values to the same as as x. Use this as a shortcut to assing the same value to multiple properties.

The widget for accessing the plot parameter mapper can be displayed by accessing the "mapper_widget" attribute of the DataFrameScatter.

```python
#Image(filename='scatter_data/mapper.png') 
ds.mapper_widget
```

In the above examples we have decided to plot the sepal_length vs the sepal width coloured by the species of each sample. This can be done by activating the fill_color property and mapping it to a column. A suitable matplotlib colormap can be selected in the free params menu.

## Free parameters

The free parameters of a plot are the visual parameters of a marker that are not mapped to a column and the colormaps used for mapped values, so each property value will be the same for all the samples. This parameters serve a default value for each attribute that is not currently mapped.

The free parameters widget can be accessed as the free_widget attribute of de ScatterDataFrame as shown below

```python
Image(filename='scatter_data/free.png') 
#ds.free_widget
```

### Tip: emphasising overlaps of sample classes

For example it is possible to activate the fill_color and line_color in the free paramers widget and then change the colormap as shown in the above cell. Selecting diverging colormaps (terrain_r for example) for the line_color property can help us understand better the overlaps among different categories, as they will be shown in a different color. Overlaps among the same class can be emphasised by change the line alpha value so the line will be darker when an overlap among samples of the same category happen.

You can inspect the overlaps hovering the mouse over a marker and reading its tooltip.

# Scale parameters

When mapping a marker attribute to a column it is possible to select the range that the mapped values will take. Currently only the maximum and minimum range of the mapping can be defined. in the next release it will be possible to select how the data values are mapped to a 0-1 range before being renormalized to a [min,max] range.

The mapping scale widget can be accessed using the attribute "scale_widget".

```python
Image(filename='scatter_data/scale.png') 
#ds.scale_widget
```
